```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.arange(1,10)
y= np.arange(11,20)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2-D')
# plt.savefig('test.png')
```

```python
plt.plot(x,y,'r--*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2-D')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'g--')
```

```python
a=np.arange(1,20,3)
b= 3*a+5
plt.plot(a,b)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Demo')
```

```python
a1= np.arange(0,4*np.pi,0.1)
b1=np.sin(a1)
plt.plot(a1,b1)
plt.title('Sine-wave')
```

```python
x=[2,8,10]
y=[3,6,9]

x1=[3,4,5]
y1=[7,9,10]

plt.bar(x,y)
plt.bar(x1,y1,color='g')
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0.05, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
```
